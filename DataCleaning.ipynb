```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('commit_data.csv')
df
```

#### 1. Xóa các dữ liệu bị thiếu

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

#### 2. Xử lý các dữ liệu trùng lặp

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

#### 3. Xoá các commit không thay đổi file nào

```python
df.drop(index=df.loc[df['Num Changed Files'] == 0].index, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

#### 4. Xử lý cột thời gian commit

```python
df['Commit Date'] = pd.to_datetime(df['Commit Date'])
df['Year']= df['Commit Date'].dt.year
df['Hour'] = df['Commit Date'].dt.hour
df
```

#### Lưu dữ liệu đã làm sạch vào 1 file csv mới để thực hiện phân tích

```python
df.to_csv("clean_commit_data.csv", index=False)
```
